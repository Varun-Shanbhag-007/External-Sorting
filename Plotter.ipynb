```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_columns(filename):
    with open(filename) as f:
        for l in f:
            if l[0] != 'D':
                continue
            else:
                print(l)
                return l.split()
        else:
            raise LookupError
def get_lines(filename, colnum):
    with open(filename) as f:
        for l in f:
            if l[0] == 's':
                yield l.split(maxsplit=colnum - 1)        

filename = 'b'
columns = read_columns(filename)
exclude = 'rrqm/s','wrqm/s','%rrqm','%wrqm', 'r_await','w_await','aqu-sz','rareq-sz','wareq-sz','svctm','%util', 
df = pd.DataFrame.from_records(
    get_lines(filename, len(columns)), columns=columns, exclude=exclude
)

print(df)
df['r/s'] = df['r/s'].astype(str).astype(float)
df['w/s'] = df['w/s'].astype(str).astype(float)
print(len(df['r/s']))

for X in df['r/s'] :
    print(X)
    
for X in df['w/s'] :
    print(X)

df.info()
```

```python
fig, axes = plt.subplots(len(df.PID.unique()), 2, figsize=(12, 8))
x_range = [df.index.min(), df.index.max()]
for i, pid in enumerate(df.PID.unique()):
    subdf = df[df.PID == pid]
    title = ', '.join([f'PID {pid}', str(subdf.index.max() - subdf.index.min())])
    for j, col in enumerate(('%CPU', 'RSS')):
        ax = subdf.plot(
            y=col, title=title if j == 0 else None, ax=axes[i][j], sharex=True
       )
        ax.legend(loc='upper right')
        ax.set_xlim(x_range)

plt.tight_layout()
plt.show()
```
